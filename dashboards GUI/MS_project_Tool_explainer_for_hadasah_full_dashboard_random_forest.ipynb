```python
!pip install explainerdashboard

import gdown
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from explainerdashboard import InlineExplainer
from explainerdashboard import ClassifierExplainer, ExplainerDashboard

url = 'https://drive.google.com/uc?id=1vaR7djiprzyQCxu5TyCSl_kpZW219eHQ' 
output = 'DF_all_treated.pkl' 
gdown.download(url, output, quiet=False)
df_all_treated = pd.read_pickle("/content/DF_all_treated.pkl")
df_all_treated
numeric_features = ['TIM3_RQ', 'TIGIT_RQ','LAG3_RQ','age_at_diagnosis','MRI_lesion_mass','Lymph_Cells','PMN_Cells','Cells']
categorical_features = ['gender','oligoclonal_band','diagnosis']
df_all_treated = pd.get_dummies(df_all_treated, columns=categorical_features, drop_first=True)
df_all_treated.drop(['treatment_when_sample_was_taken'], axis='columns', inplace=True)
df_all_treated.columns.values
df_all_treated[df_all_treated['diagnosis_SPMS']==0].reset_index()
predictor_var = ['sample_code', 'TIM3_RQ', 'TIGIT_RQ', 'LAG3_RQ','age_at_diagnosis', 'primary_EDSS_at_diagnosis', 'MRI_lesion_mass','Lymph_Cells','PMN_Cells','Cells','gender_M','oligoclonal_band_POS']
outcome_var = 'diagnosis_SPMS'
X_train, X_test, y_train, y_test = train_test_split(df_all_treated[predictor_var], df_all_treated[outcome_var])
test_names = X_test["sample_code"]
X_train.drop(['sample_code'], axis='columns', inplace=True)
X_test.drop(['sample_code'], axis='columns', inplace=True)
model = RandomForestClassifier(n_estimators=50, max_depth=10).fit(X_train, y_train)
explainer = ClassifierExplainer(model, X_test, y_test,idxs=test_names,
                                target = 'diagnosis_SPMS',
                                index_name="sample_code",
                               labels=['RRMS', 'SPMS'])
ExplainerDashboard(explainer).run()

```
